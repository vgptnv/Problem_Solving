```python
# Q. 7490
import itertools

n = int(input())
nums = list(map(int, input().split()))
op1, op2, op3, op4 = map(int, input().split())
ops = ["+"] * op1 + ["-"] * op2 + ["*"] * op3 + ["/"] * op4

operations = list(itertools.permutaions(ops, n-1))
result = []
for operation in operations:
    formula = []
    for i in range(n):
        formula.append(nums[i])
        if i == n-1:
             continue
        formula.append(operation[i])

    result.append(eval("".join(map(str, formula))))
    
result
```
