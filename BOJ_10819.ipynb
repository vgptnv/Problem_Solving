```python
# Q. 10819
from itertools import permutations 
n = int(input())
lst = list(map(int, input().split()))

maximum = 0
for perm in set(permutations(lst, n)):
    print(perm)
    total = 0
    for i in range(len(perm)-1):  # n-2까지 돌아가게
        total += abs(perm[i] - perm[i+1])

    maximum = max(maximum, total)

print(maximum)
```

```python
from itertools import permutations

n=int(input())
arr = list(map(int,input().split()))

candidate = list(permutations(arr,r=n))

ans =-1

def cal(t):
    res =0
    for i in range(len(t)-1):
        temp = abs(t[i]-t[i+1])
        res+= temp
    return res

for i in candidate:
    ans = max(ans,cal(i))


print(ans)
```

```python
# Q. 14889
from itertools import permutations
n, m = map(int, input().split())
lst = map(int, input().split())

maximum = 0
for perm in set(permutations(lst, 3)):
    if sum(perm) >= maximum and sum(perm) <= m:
        maximum = sum(perm)
print(maximum)
```

```python
from itertools import combinations  # 조합 함수

N = int(input())
S = [list(map(int, input().split())) for _ in range(N)]
members = [i for i in range(N)]  # n=4, [0, 1, 2, 3]
possible_team = []

# 조합으로 가능한 팀 생성해주기
for team in list(combinations(members, N//2)):
    possible_team.append(team)

min_stat_gap = 10000 #갭이 가장 작은 값을 찾기 위하여
for i in range(len(possible_team)//2):
    #A 팀
    team = possible_team[i]  # ex. (0, 1) -> 두 명의 조합을, (0, 1) (1, 0) 다르게
    stat_A = 0 #A팀 능력치
    # 이중 반복문
    for j in team:  # 한 팀의 멤버 수
        member = j # 멤버
        for k in team:
            stat_A += S[member][k]  # 멤버와 함께할 경우의 능력치들
        # S[0][0], S[0][1], S[1][0], S[1][1]
    
    # possible_team = [(0, 1), (0, 2), (0, 3), (1, 2), (1, 3), (2, 3)]
    #A를 제외한 나머지 팀 
    team = possible_team[-i-1]
    stat_B = 0
    for j in range(N//2):
        member = team[j]
        for k in team:
            stat_B += S[member][k]
            
    min_stat_gap = min(min_stat_gap, abs(stat_A - stat_B))  # abs : 절댓값함수
    
print(min_stat_gap)
```
