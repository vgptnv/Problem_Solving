```python
# Q. 7490
import itertools

n = int(input())

for _ in range(n):
    num = int(input())
    lst = [i for i in range(1, num+1)]
    operations = list(itertools.product(["+", "-", ""], repeat=num-1))
    operations.sort()
    for operation in operations:
        formula = []
        for i in range(num):
            formula.append(lst[i])
            if i == num-1:
                 continue
            formula.append(operation[i])
        
        if eval("".join(map(str, formula))) == 0:
            print("".join(map(str, [" " if x=="" else x for x in formula])))
    print()
```
