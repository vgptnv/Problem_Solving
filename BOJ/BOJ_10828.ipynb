```python
# 10828. 스택
import sys
n = int(sys.stdin.readline())

stack = []

def push(x):
    stack.append(x)

def pop():
    if len(stack) == 0:
        return -1
    else :
        return stack.pop()

def size():
    return len(stack)

def empty():
    if len(stack)==0:
        return 1
    else:
        return 0

def top():
    if len(stack) == 0:
        return -1
    else:
        return stack[-1]
    
for _ in range(n):
    result = sys.stdin.readline().split()
    command = result[0]
    
    if command == "push":
        push(result[1])
    elif command == "pop":
        print(pop())
    elif command == "size":
        print(size())
    elif command == "top":
        print(top())
    elif command == "empty":
        print(empty())
    else:
        print('input command error')
        break
```
