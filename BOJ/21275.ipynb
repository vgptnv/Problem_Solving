```python
# Q. 21275
a, b = input().split()

def convert_to_num_lst(number):
    result = []
    for i in number:
        if i.isdigit():
            result.append(int(i))
        else:
            result.append(ord(i) - 87)

    return result

def find_cases(lst):
    result = [0 for i in range(37)]
    lst_inv = lst[::-1]
    for digit in range(len(lst)):  # 자릿수
        for i in range(2, 37):  # 진법 (가능한 a, b의 값 후보들)
            if lst_inv[digit] < i:
                result[i] += lst_inv[digit] * (i ** digit)
                
    return result

a_lst = convert_to_num_lst(a)
b_lst = convert_to_num_lst(b)

a_cases = find_cases(a_lst)
b_cases = find_cases(b_lst)

result = []
for i in range(2, 37):
    if a_cases[i] != 0 and a_cases[i] in b_cases[2:]:
        if i != b_cases.index(a_cases[i]):
            result.append((a_cases[i], i, b_cases.index(a_cases[i])))  # x, a, b

if len(result) == 0:
    print('Imposibble')
elif len(result) == 1:
    print(*result[0])
else:
    print('Multiple')
```

```python
# 0. 입력받기
import sys
A, B = input().split()
x, a, b = 0, 0, 0 # 출력할 값
count = 0 # 조건에 맞는 값이 몇개인지 카운팅해주는 변수


# 1. A를 가지고 2부터 36진수 중 표현이 가능한 값을 찾음
for i in range(2, 37):
    try:
        a_temp = int(A, i)
    except:
        continue
    # 2. B를 가지고 2부터 36진수 중 표현이 가능한 값을 찾음
    for j in range(2, 37):
        try:
            b_temp = int(B, j)
            # 3. 진수 변환한 값들이 같을 떄
            if a_temp == b_temp:
                count += 1
                x = a_temp
                a = i
                b = j
        except:
            continue

# 4. 출력하기
if count == 1:
    print(x,a,b)
elif count == 0:
    print("Impossible")
else:
    print("Multiple")
```
